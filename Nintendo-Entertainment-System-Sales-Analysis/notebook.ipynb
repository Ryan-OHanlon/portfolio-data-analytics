# Nintendo Entertainment System Sales Analysis

**Background**

The Nintendo Entertainment System, also known as the Famicom in Japan. The hobby of video games would probably of ended if it wasn't for Nintendo developing their own console and reviving the video game industry after the video game crash.

The focus of this analysis is a deep dive into the highest selling games for the NES. We'll be looking at how many games sold the most units, what genre sold the most, and what companies were involved in publishing these games. We'll also compare how these popular video games performed across the North American region, European region, Japan, and other parts of the world to find out what game genres each region preferred, which publisher sold the most units across each region and which game was the most popular by release year.

```r
suppressPackageStartupMessages(library(tidyverse))

read_csv('data/vgsales.csv.gz', show_col_types = FALSE)
```

## Data Dictionary

**The Data**

This dataset contains records of popular video games in North America, Japan, Europe and other parts of the world. Every video game in this dataset has at least 100k global sales.

| Column        | Explanation                                                                   |
| ------------- | ----------------------------------------------------------------------------- |
| Rank          | Ranking of overall sales                                                      |
| Name          | Name of the game                                                              |
| Platform      | Platform of the games release (i.e. PC,PS4, etc.)                             |
| Year          | Year the game was released in                                                 |
| Genre         | Genre of the game                                                             |
| Publisher     | Publisher of the game                                                         |
| NA_Sales      | Number of sales in North America (in millions)                                |
| EU_Sales      | Number of sales in Europe (in millions)                                       |
| JP_Sales      | Number of sales in Japan (in millions)                                        |
| Other_Sales   | Number of sales in other parts of the world (in millions)                     |
| Global_Sales  | Number of total sales (in millions)                                           |

[Source](https://www.kaggle.com/gregorut/videogamesales) of dataset.

Since the focus of the analysis is on the Nintendo Entertainment System, the dataset `Game_Sales` will be filtered to create the variable `NES_Game_Sales`.

```r
library("tidyverse")
```

```r
Game_Sales <- read_csv('data/vgsales.csv.gz')
```

```r
NES_Game_Sales <- Game_Sales %>%
	filter(Platform == "NES")

NES_Game_Sales
```

# How many NES games sold over 100K units?

```r
NES_Game_Sales %>%
	nrow()
```

## Which genre had the most successful games for the NES?

```r
NES_Game_Sales %>%
	ggplot(aes(reorder(Genre, Genre, function(x) length(x)))) +
	geom_bar() +
    theme(axis.text.x = element_text(angle = 90)) +
	labs(title = "Games that sold 100k units by Genre") +
    xlab("Genre") +
    ylab("Number of games")
```

## Which publisher released the most popular games for the NES?

```r
NES_Game_Sales %>%
	ggplot(aes(reorder(Publisher, Publisher, function(x) length(x)))) +
	geom_bar() +
	theme(axis.text.x = element_text(angle = 90)) +
    labs(title = "Games that sold 100k units by Publisher") +
    xlab("Publisher") +
    ylab("Number of games")
```

## Which release year had the most popular games for the NES?

```r
#Games that sold over 100k by game release year
NES_Game_Sales %>%
	group_by(Year) %>%
	summarise(Count = n()) %>%
	ggplot(aes(Year, Count))+
	geom_col() +
	labs(title = "Games that sold 100k units by year") +
    ylab("Number of games")
```

```r
NES_Game_Sales %>%
	select(-Rank) %>%
	filter(Year == 1986) %>%
	arrange(desc(Global_Sales))
```

Out of the 716 licensed games that were released globally, ninty-eight games sold more than 100k units. 

The most popular genre for the NES were platformer games with 28 games breaking 100k units sold. Followed by sports, puzzle, action and role-playing games each having 10 games that have become stable genres in the video game industry.

The number one publisher for the NES was Nintendo with 45 games being the most popular games. As the years went on during the NES lifespan, other companies that had their games originally published by Nintendo struck it out on their own moving from being a developer to a publisher. Companies such as Namco Bandai, SquareSoft/Square, Enix, Konami, and Capcom are all still around and are still known as some of the biggest video game publishers today.

What is interesting is that games released in 1986 had 19 games that sold more than 100k units. Games like The Legend of Zelda, World Class Track Meet, and Metroid really brought the NES into the mainstream.

# Region Total Sales

```r
NES_Region_Sales <- NES_Game_Sales %>%
	pivot_longer(c("NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"), names_to = "Region", values_to = "Sales")
```

## Which region had the most sales?

```r
NES_Region_Sales %>%
	group_by(Region) %>%
	summarise(total_sales = sum(Sales)) %>%
	arrange(desc(total_sales))

NES_Region_Sales %>%
	group_by(Region) %>%
	summarise(total_sales = sum(Sales)) %>%
	filter(total_sales != 0) %>%
	ggplot(aes(reorder(Region, total_sales), total_sales, fill = Region)) +
	geom_col() +
	guides(fill = FALSE) +
	ylab("Total Unit Sales (in millions)")+
	xlab("Region")+
	labs(title = "Region Total Unit Sales")+
	scale_x_discrete(label = c("Other", "Europe", "Japan", "North America"))
```

## What was the most popular genre by region?

```r
NES_Region_Sales %>%
	group_by(Region, Genre) %>%
	summarise(total_sales = sum(Sales)) %>%
	ggplot(aes(reorder(Genre, total_sales), total_sales, fill = Region, group = total_sales)) +
	geom_col(position = "dodge2") +
	theme(axis.text.x = element_text(angle = 90))+
	labs(title = "Total Unit Sales by Genre")+
	xlab("Genre")+
	ylab("Total Unit Sales (in millions)")+
	scale_fill_discrete(labels = c("Europe", "Japan", "North America", "Other"))
```

## Which Publisher sold the most units in each region?

```r
NES_Region_Sales %>%
	group_by(Region, Publisher) %>%
	summarise(total_sales = sum(Sales)) %>%
	ggplot(aes(reorder(Publisher, total_sales), total_sales, fill = Region, group = total_sales)) +
	geom_col(position = "dodge2") +
	theme(axis.text.x = element_text(angle = 90))+
	labs(title = "Total Unit Sales by Publisher")+
	xlab("Publisher")+
	ylab("Total Unit Sales (in millions)")+
	scale_fill_discrete(labels = c("Europe", "Japan", "North America", "Other"))
```

## How did sales perform in each region over the years?

```r
NES_Region_Sales %>%
	group_by(Region, Year) %>%
	summarise(total_sales = sum(Sales)) %>%
	ggplot(aes(Year, total_sales, fill = Region, group = total_sales)) +
	geom_col(position = "dodge2")+
	labs(title = "Total Unit Sales by Release Year")+
	xlab("Release Year")+
	ylab("Total Unit Sales (in millions)")+
	scale_fill_discrete(labels = c("Europe", "Japan", "North America", "Other"))
```

Surprisingly, despite the video game crash, the North American region had the most sales with 125 million units sold followed by the Japan at 98 million sales. Interesting of note is the European region with only 21 million total sales while remaining regions had 5 million sales.

Looking at the most popular genres by region shows that each region preferred the platform genre the most. The North American region sold over 40 million units in the platform genre, Japan sold over 25 million units, and the European region sold over 10 million units. Outside of platformers, each region showed different preferences. The North American region preferred shooters and action games. The Japan region preferred sports and role-playing games over action and shooters. The European region preferred action and sports games.

Nintendo being the primary publisher for the Nintendo Entertainment System showed how much market dominance they had across all regions with over 95 million games sold in the North American region alone. However, as more developers started to become publishers we can see how other companies started to emerge. Capcom performed well across all regions, Namco Bandai, Enix, Hudson Soft and Squaresoft performed well in Japan.

During the lifetime of the NES the peak years of the NES in terms of sales started in 1984 and ended in 1988. The North American region made over half of the total sales during 1984 and 1985 alone with over 30 million games being sold. Japan sales was the most consistent with almost 20 million games sold during 1986. The European region had the most sales during 1985 and 1988, while other regions had the most sales during 1986.
